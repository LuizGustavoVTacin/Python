#### Trabalhando com datas temporais:

Trabalhar com data, geralemente, acontece com adição de valores temporais no eixo x.

Para isso, é necessário trabalhar com as bibliotecas datetime e matplotlib.dates 

O uso destas bibliotecas permite a criação de escalas de tempo, bem como alterar o tipo de exibição temporal.

As escalas de tempo podem ser criadas com: DayLocator(), MonthLocator() e YearLocator.                      
A exibição temporal é feita com a função DateFormatter().

Após definir os modelos temporais, pode-se definir o eixo x, de acordo com a hierarquia com a funções set_major_locator() e set_minor_locator() e utilizar a exibição temporal criada com set_major_formatter().

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

months = mdates.MonthLocator()
days = mdates.DayLocator()

timeFmt = mdates.DateFormatter('%Y-%m')
events = [datetime.date(2015,1,23), datetime.date(2015,1,28),
          datetime.date(2015,2,3), datetime.date(2015,2,21),
          datetime.date(2015,3,15), datetime.date(2015,3,24),
          datetime.date(2015,4,8),
          datetime.date(2015,4,24)]
readings = [12, 22, 25, 20, 18, 15, 17, 14]

fig, ax = plt.subplots()
plt.plot(events, readings)
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
plt.show()
```

#### Gráfico de Barras:

Gráfico de barras são semelhantes aos histogramas, mas com separações entre eles e referenciados por categorias, como, por exemplo, cada barra representando uma a frequência de ocorrência de uma cor.

Gráfico de barras são construídos com a função plt.bar()

```python
import matplotlib.pyplot as plt

index = np.arange(5)
valores = [3, 7, 5, 1, 2]
plt.bar(index, valores)
plt.xticks(index,['Amarelo','Azul','Verde','Vermelho','Preto'])
plt.show()
```

Gráficos de barras podem ser personalizados para adicionar informações adicionais, por meio dos kwargs.

Uma destas personalizações, é a adição das barras de erro.

A função que adiciona barras de erro é a error_kw.
No exemplo abaixo, o erro é obtido pelo desvio padrão std.

O parâmetro alpha indica a transparência da barra colorida, contido entre 0 e 1 e, por fim, há o capsize que indica o comprimento da linha transversal da barra de erro.

```python
import numpy as np
import matplotlib.pyplot as plt

index = np.arange(5)
valores = [3, 7, 5, 1, 2]
std = [0.8, 1, 0.4, 0.25, 0.3]
plt.bar(index, valores, yerr = std, error_kw={'ecolor':'0.1','capsize':6}, alpha=0.7)
plt.xticks(index,['Amarelo','Azul','Verde','Vermelho','Preto'])
plt.show()
```

#### Gráfico de barras multi-serial:

O gráfico de barras multi-serial consite em mais de 1 barra representar cada categoria.

A criação deste gráficos consiste em dividir o espaço de 1 index no número de gráficos que compõem cada categoria.

Este parâmetro de divisão é uma variável declarada no código, no caso bw.

```python
import matplotlib.pyplot as plt
import numpy as np

index = np.arange(5)
valores1 = [5, 7, 3, 4, 6]
valores2 = [6, 6, 4, 5, 7]
valores3 = [5, 6, 5, 4, 6]
bw = 0.3
plt.axis([0,5,0,8])
plt.bar(index,valores1, bw, color='b')
plt.bar(index+bw, valores2, bw, color='g')
plt.bar(index+2*bw, valores3, bw, color='r')
plt.xticks(index+1.5*bw, ['A','B','C','D','E'])
plt.show()
```

#### Gráfico de Pizza:

Gráfico de pizza é construído com a função plt.pie()

Os dados inseridos podem ser porcentagens ou valores comuns. Para plotar o gráfico como uma esfera simétrica, usar 'equal' como argumento da função plt.axis()

```python
import matplotlib.pyplot as plt

labels=['Nokia','Samsung','Apple','Lumia']
valores = [10,30,45,15]
colors = ['yellow','green','red','blue']
plt.pie(valores,labels=labels,colors=colors)
plt.axis('equal')
plt.show()
```

Uma modificação possível para este gráfico é a plotagem de uma parte saindo da esfera.

Isto é obtido pelo parâmetro explode=. especificando o ângulo de início da esfera pelo parâmetro startangle. 

A função explode recebe valores entre 0 e 1, tal que 0 significa completamente dentro da esfera e 1, completamente fora. Explode receberá valores para cada parte do gráfico, de forma que será atribuído o valor na ordem em que a parte foi declarada na label.

```python
import matplotlib.pyplot as plt

labels=['Nokia','Samsung','Apple','Lumia']
valores = [10,30,45,15]
colors = ['yellow','green','red','blue']
explode = [0.3, 0, 0, 0]
plt.pie(valores,labels=labels,colors=colors, explode = explode, startangle=180)
plt.axis('equal')
plt.show()
```

Para adicionar a porcentagem correspondente a cada parte, usa-se o parâmetro autopct

```python
import matplotlib.pyplot as plt 

labels = ['Nokia','Samsung','Apple','Lumia']
valores = [10,30,45,15]
colors=['yellow','green','red','blue']
explode = [0.3,0,0,0]
plt.pie(valores,labels=labels,colors=colors,explode=explode,shadow=True,autopct='%1.1f%%', startangle=180)
plt.axis('equal')
plt.show()
```

#### Gráficos de Contorno (Curvas de nível):

Um gráfico de contorno é gerado por uma função de duas variáveis pelo comando contour()

```python
import matplotlib.pyplot as plt
import numpy as np 

def f(x,y):
    return (1-y**5 + x**5)*np.exp(-x**2-y**2)
dx = 0.01; dy = 0.01
x = np.arange(-2,2,dx)
y = np.arange(-2,2,dy)
X,Y = np.meshgrid(x,y)
C = plt.contour(X,Y,f(X,Y),8,colors='black')
plt.contourf(X,Y,f(X,Y),8)
plt.clabel(C, inline=1,fontsize=10)
plt.show()
```

No gráfico acima, a função meshgrid criará um retângulo com pontos calculadas em todos os valores de x multiplicados pelos pontos de y.

A função contour() gerará curvas de nível com linhas sólidas e a função contourf() gerará curvas de nível com áreas preenchidas. Abaixo, há gráficos com ambas funções utilizadas. 

Por fim, a função plt.clabel é utilizada para colocar rótulos a cada nível das curvas

```python
import matplotlib.pyplot as plt
import numpy as np 

def f(x,y):
    return (1-y**5 + x**5)*np.exp(-x**2-y**2)

fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True)
dx = 0.01; dy = 0.01
x = np.arange(-2,2,dx)
y = np.arange(-2,2,dy)
X,Y = np.meshgrid(x,y)

ax1.contour(X,Y,f(X,Y),8,colors='black')
ax1.clabel(ax1.contour(X,Y,f(X,Y),8,colors='black'), inline=1,fontsize=10)

ax2.contourf(X,Y,f(X,Y),8)
plt.show()
```
